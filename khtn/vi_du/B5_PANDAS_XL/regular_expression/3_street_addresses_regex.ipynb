```python
import re
import pandas as pd
```

```python
# rút trích số nhà từ địa chỉ
f = open('street_addresses.txt')
for line in f:
    line = line.strip()
    regex = r"[0-9]+"
    match = re.search(regex, line)
    if match:
        print(line, "=>", match.group(0))
f.close()
```

```python
# rút trích tên đường từ địa chỉ
f = open('street_addresses.txt')
for line in f:
    line = line.strip()
    regex = r"([a-zA-Z]|\s|[-])+"
    match = re.search(regex, line)
    if match:
        print(line, "=>", match.group(0))
f.close()
```

```python
# doc tap tin street_addresses.txt va tao df de luu
f = open('street_addresses.txt')
l = []
for line in f:
    line = line.strip()
    l.append(line)
f.close()    
```

```python
df = pd.DataFrame({'address':l})
```

```python
df.tail()
```

```python
def get_house_no(line):
    regex = r"[0-9]+"
    match = re.search(regex, line)
    if match:
        return match.group(0)
    else:
        return None
```

```python
df['house_no'] = df['address'].map(lambda line: get_house_no(line))
```

```python
df
```

```python
def get_house_street(line):
    regex = r"([a-zA-Z]|\s|[-])+"
    match = re.search(regex, line)
    if match:
        return match.group(0)
    else:
        return None
```

```python
df['house_street'] = df['address'].map(lambda line: get_house_street(line))
```

```python
df['no'] = df['address'].str.extract('([0-9]+)')
```

```python
df['street'] = df['address'].str.extract('(([a-zA-Z]|[ ]|[-])+)')[0]
```
