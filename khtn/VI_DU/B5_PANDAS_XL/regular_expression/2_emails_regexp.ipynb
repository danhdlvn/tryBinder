```python
import re
```

```python
# rút trích tất cả email
f = open('emails.txt')
for line in f:
    line = line.strip()
    regex = r"[a-zA-Z]\S*@\S*[a-zA-Z]"
    matches = re.findall(regex, line)
    if len(matches)>0:
        print(matches)
f.close()
```

```python
# rút trích email của From:
f = open('emails.txt')
for line in f:
    line = line.strip()
    regex = r"^From: [a-zA-Z]\S*@\S*[a-zA-Z]"
    matches = re.findall(regex, line)
    if len(matches)>0:
        print(matches)
f.close()
```

```python
# rút trích email kết thúc là .org
f = open('emails.txt')
for line in f:
    line = line.strip()
    regex = r"[a-zA-Z]\S*@\S*[a-zA-Z].org$"
    matches = re.findall(regex, line)
    if len(matches)>0:
        print(matches)
f.close()
```

```python
# rút trích email kết thúc là .org hoặc .edu
f = open('emails.txt')
for line in f:
    line = line.strip()
    regex = r"[a-zA-Z]\S*@\S*[a-zA-Z].[org|edu]$"
    matches = re.findall(regex, line)
    if len(matches)>0:
        print(line,"=>",matches)
f.close()
```
